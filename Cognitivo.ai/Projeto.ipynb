### Carregando dataset

### Analise exploratória

```r
#Distribuição de tipo de quartos

df.group <- dataset %>%
            group_by(.data$room_type) %>%
            summarise(freq = n(),.groups = 'drop') %>% 
            arrange(desc(freq))  %>%
            mutate(per = paste0(round(freq / sum(freq)  * 100,2),"%"))

#Tipo de quarto x frequencia
df.group %>% 
    ggplot(aes(x = room_type,y = freq)) + 
    geom_bar(stat = 'identity', position = ggplot2::position_stack()) + 
    geom_text(ggplot2::aes(label = per),position = ggplot2::position_stack(),size = 3,check_overlap = TRUE,hjust = -0.2)  +
    coord_flip() +
    ggtitle("Distribuição dos tipos de quartos")

```

```r
#distribuição da vizinhança top 10
df.group <- dataset %>%
  group_by(.data$neighbourhood) %>%
  summarise(freq = n(),.groups = 'drop') %>% 
  arrange(desc(freq))  %>%
  mutate(per = paste0(round(freq / sum(freq)  * 100,2),"%")) %>%
  top_n(10)

#Tipo de quarto x frequencia
df.group %>% 
  ggplot(aes(x = neighbourhood,y = freq)) + 
  geom_bar(stat = 'identity', position = ggplot2::position_stack()) + 
  geom_text(ggplot2::aes(label = per),position = ggplot2::position_stack(),size = 3,check_overlap = TRUE,hjust = -0.2)  +
  coord_flip() +
  ggtitle("Top 10 da distribuição da vizinhança")
```

## Classificação do room type (feature ‘room_type’)

```r
#o modelo_pca filter enginer 
#o dataset normal
df                     <- dataset %>% drop_na()
df$latitude            <- NULL
df$longitude           <- NULL
df$last_review         <- NULL
df$name                <- NULL
df$id                  <- NULL
df$name                <- NULL
df$host_id             <- NULL
df$host_name           <- NULL

#transform caracter em fator
df <- df %>% 
      mutate_if(is.character,function(x) as.factor(x))
 
df$neighbourhood <- as.numeric(df$neighbourhood)
                
#Usando random forest para obter as  variaveis mais importantes para classificação do model                
model_pca <- randomForest(room_type  ~ .,data = df,ntree=500, keep.forest=FALSE)
varImpPlot(model_pca)
```

```r
#Existe um grande balanceamento das classes, vamos aplicar o metedo de normalização dos dados para corregir isso

table(df$room_type)
```

```r
#o modelo_pca filter enginer 
#o dataset normalizado 
df.scale                   <- dataset %>% drop_na()
df.scale$price             <- sapply(df.scale$price,function(x) (x - mean(df.scale$price)) / sd(df.scale$price))
df.scale$minimum_nights    <- sapply(df.scale$minimum_nights,function(x) (x - mean(df.scale$minimum_nights)) / sd(df.scale$minimum_nights))
df.scale$number_of_reviews <- sapply(df.scale$number_of_reviews,function(x) (x - mean(df.scale$number_of_reviews)) / sd(df.scale$number_of_reviews))
df.scale$reviews_per_month <- sapply(df.scale$reviews_per_month,function(x) (x - mean(df.scale$reviews_per_month)) / sd(df.scale$reviews_per_month))
df.scale$calculated_host_listings_count <- sapply(df.scale$calculated_host_listings_count,function(x) (x - mean(df.scale$calculated_host_listings_count)) / sd(df.scale$calculated_host_listings_count))
df.scale$availability_365  <- sapply(df.scale$availability_365,function(x) (x - mean(df.scale$availability_365)) / sd(df.scale$availability_365))

#Removendo as variáveis que não vão ser uteis para esse modelo
df.scale$latitude            <- NULL
df.scale$longitude           <- NULL
df.scale$last_review         <- NULL
df.scale$name                <- NULL
df.scale$id                  <- NULL
df.scale$name                <- NULL
df.scale$host_id             <- NULL
df.scale$host_name           <- NULL

#transforma caracter em fator
df.scale <- df.scale %>% mutate_if(is.character,function(x) as.factor(x))

df.scale$neighbourhood <- as.numeric(df.scale$neighbourhood)

#Usando random forest para obter as  variaveis mais importantes para classificação do model                
model_pca_normalizado <- randomForest(room_type  ~ .,data = df.scale,ntree=500, keep.forest=FALSE)
varImpPlot(model_pca_normalizado)
```

### Modelo NaiveBayes com dados normais

```r
#Vimos podemos afirmar que são 4 [PRICE,REVIEWS_PER_MONTH,AVAILABILITY_365,NEIGHBOURHOOD] variais necessária para o modelo de classificação de room_type

#Dataset com dados normais sem scale aplicada

#obter os index no qual vamos separa o trainamento dos teste senso que 75% será trainado e 25% sera para testes
amostra <- sample(x = 2,size = nrow(df),replace = T,prob = c(0.75,0.25))

#separando o dataset e obtendo as variaveis mais principais para o modelo de classificação
df.train <- df[amostra == 1,c("price","reviews_per_month","availability_365","neighbourhood","room_type")]
df.test  <- df[amostra == 2,c("price","reviews_per_month","availability_365","neighbourhood","room_type")]

#modelo probabilidade bayes
model <- naiveBayes(room_type ~ .,data = df.train,laplace = 1)

#prevendo a classificação dos previsores teste
previsao <- predict(model,newdata = df.test)

#Matrix de confusao
validacao <- table(previsao,df.test$room_type)
confusionMatrix(validacao)
```

### Modelo NaiveBayes com dados normalizado

```r
#Dataset com dados normais com scale aplicada

#obter os index no qual vamos separa o trainamento dos teste senso que 75% será trainado e 25% sera para testes
amostra <- sample(x = 2,size = nrow(df.scale),replace = T,prob = c(0.75,0.25))

#separando o dataset e obtendo as variaveis mais principais para o modelo de classificação
df.train <- df.scale[amostra == 1,c("price","reviews_per_month","availability_365","neighbourhood","room_type")]
df.test  <- df.scale[amostra == 2,c("price","reviews_per_month","availability_365","neighbourhood","room_type")]

#modelo probabilidade bayes
model <- naiveBayes(room_type ~ .,data = df.train,laplace = 2)

#prevendo a classificação dos previsores teste
previsao <- predict(model,newdata = df.test)

#Matrix de confusao
validacao <- table(previsao,df.test$room_type)
confusionMatrix(validacao)
```

## Modelo k-Nearest Neighbour com dados normais

```r
#Dataset com dados normais sem scale aplicada

#obter os index no qual vamos separa o trainamento dos teste senso que 75% será trainado e 25% sera para testes
amostra <- sample(x = 2,size = nrow(df),replace = T,prob = c(0.75,0.25))

#separando o dataset e obtendo as variaveis mais principais para o modelo de classificação
df.train <- df[amostra == 1,c("price","reviews_per_month","availability_365","neighbourhood","room_type")]
df.test  <- df[amostra == 2,c("price","reviews_per_month","availability_365","neighbourhood","room_type")]

#criando o modelo e prevendo os dados de teste
previsao <- class::knn(df.train[,1:4],df.test[,1:4],cl = df.train[,5],k = 5,prob = TRUE)
#validação do modelo
confusionMatrix(table(previsao,df.test[,5]))
```
